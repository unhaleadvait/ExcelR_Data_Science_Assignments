##             HYPOTHESIS TESTING ASSIGNMENT



**BACKGROUND**



Bombay Hospitality Ltd. operates a franchise model that produces exotic Norwegian dinners. According to their original cost formula, the weekly operating cost (W) is:



**W=1000+5XW = 1000 + 5XW=1000+5X**  



the number of units produced in a week is X. However, restaurant owners have started reporting that their actual weekly costs are higher than what this model predicts. To investigate whether there is statistical evidence to support this claim, we apply hypothesis testing using a sample of recent weekly costs.



GOAL is to determine whether, using recent sample data from 25 restaurants, the average weekly operating cost is noticeably higher than what the cost model forecasts.



**1: STATE THE HYPOTHESES STATEMENT**

1. **Null Hypothesis (H₀):** The average weekly cost does not increase as the model predicted.  
$H_0: \mu = 4000$

2. **Alternative Hypothesis (H₁):** The average weekly cost is higher than expected.  
$H_1: \mu > 4000$




 **2:      CALCULATE THE TEST STATISTIC**





Since the population standard deviation is known, we are employing the Z-test.

z-score formula: $z = \frac{\bar{X} - \mu}{\sigma / \sqrt{n}}$












Where:  
$ \bar{X} = 3050 $ (sample mean),  
$ \mu = 4000 $ (theoretical mean using the cost model with $X = 600$),  
$ \sigma = 5 \times 25 = 125 $ (cost standard deviation),  
$ n = 25 $ (sample size)
  

```python
import math


# Given data
x_bar = 3050          # Sample mean
mu = 4000             # Theoretical mean
sigma = 125           # Standard deviation
n = 25                # Sample size

# Calculating z-score
z = (x_bar - mu) / (sigma / math.sqrt(n))
print(f"Z-test statistic: {z:.2f}")
```

This indicates that z = -38.00 is the test statistic so we will then compare  

to the critical value.

**3: DECIDING THE CRITICAL VALUE**

Significance level: $ \alpha = 0.05 $  

Since this is a one-tailed test (looking for an increase), we use:  
$ Z_{critical} = 1.645 $


**4. Verifying whether the outcome validates the claim**



In this step we will compare the calculated z-value with the critical value taken from the z-table in order to find out whether the restaurant owners' claim is supported by sufficient evidence.

```python

# CHECKING IF THE RESULT SUPPORTS THE CLAIM

if z > 1.645:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
```

We fail to disprove the null hypothesis since the test statistic (-7.60) is notably less than the critical value (1.645).  

This result points to the hypothesis that weekly operating costs have not been increased and hence, the sample data offer no evidence in support of this assertion.







**CONCLUSION**

We are unable to rule out the null hypothesis based on the data and hypothesis test. This indicates that there isn't any statistical proof that the average weekly operating cost has gone up above what the model estimates.
