**#          1 PRIME NUMBERS**

goal is to figure out if the number in question is prime

```python
# This function checks if a number is a prime number.
# A prime number is greater than 1 and has no divisors other than 1 and itself.
def is_prime(n):
    if n <= 1:
        return False  # Numbers 1 and below are not prime
    # Loop through numbers from 2 up to the square root of n
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False  # If divisible, itâ€™s not prime
    return True  # If no divisors found, the number is prime

# Example check
number = 17
print(f"{number} is prime: {is_prime(number)}")
```

**2.     PRODUCT OF RANDOM NUMBERS**



goal is to create two random numbers and check if the user's input is correct.

```python
# This program generates two random numbers and asks the user to multiply them
import random

# Randomly pick two integers between 1 and 10
a = random.randint(1, 10)
b = random.randint(1, 10)

# Display the numbers and prompt the user to enter their product
print(f"What is the product of {a} and {b}?")
user_input = int(input("Your answer: "))

# Compare user's answer with the correct answer and give feedback
if user_input == a * b:
    print("Correct!")
else:
    print(f"Wrong! The correct answer is {a * b}")
```

besides creating two random numbers, the program correctly checks if the user's inputed product is correct

**EXERCISE 3     SQUARES OF EVEN NUMBERS (100 TO 200)**



goal is printing squares for every even number between 100 and 200.

```python
# This loop goes through numbers from 100 to 200 and prints squares of even numbers
for i in range(100, 201):
    if i % 2 == 0:            # Check if the number is even
        print(f"{i}^2 = {i**2}") # Print the number and its square

```

the given code prints the square of the even number in the range 100 to 200.

**EXERCISE 4   WORD COUNTER**



goal is to find the occurrence of each word in the given text.

```python
# This program counts how many times each word appears in a given sentence
from collections import Counter

# A sample string to work with
input_text = "This is a sample text. This text will be used to demonstrate the word counter."

# Convert to lowercase and remove the period to clean the text
words = input_text.lower().replace(".", "").split()

# Count each word using Counter
word_count = Counter(words)

# Print each word and its count
for word, count in word_count.items():
    print(f"'{word}': {count}")
```

the program gets the input text and efficiently figures out the number of times each word appears.

**EXERCISE 5    CHECKING FOR PALINDROME**



goal is without taking into account punctuation, case, or spaces, check if the string is a palindrome.

```python
# This function checks if a string is a palindrome
# It ignores spaces,punctuation,capitalization
import re

def is_palindrome(s):
    # Clean the string: remove non-letter characters and make lowercase
    cleaned = re.sub(r'[^A-Za-z0-9]', '', s).lower()
    # Compare cleaned string to its reverse
    return cleaned == cleaned[::-1]

# Testing the function with a sample input
input_str = "Racecar"
print(f"'{input_str}' is palindrome: {is_palindrome(input_str)}")


```

this function finds if the given string is a palindrome and gives the correct result after removing spaces, punctuation, and case.
